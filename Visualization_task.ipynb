```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\KAVIPRIYA\Downloads\CodeAlpha_WebScraping\books_data_inr.csv")
```

```python
df.head()
```

Convert Rating to Numeric

```python
rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

df["rating_num"] = df["Rating"].map(rating_map)
```

Price Distribution

```python
plt.hist(df["Price (INR ₹)"])
plt.title("Distribution of Book Prices")
plt.xlabel("Price in INR")
plt.ylabel("Number of Books")
plt.show()
```

Rating Distribution

```python
df["Rating"].value_counts().plot(kind="bar")
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

Average Price by Rating

```python
plt.plot(
    df.groupby("rating_num")["Price (INR ₹)"].mean()
)
plt.title("Average Price by Rating")
plt.xlabel("Rating")
plt.ylabel("Average Price")
plt.show()
```

Availability Count

```python
df["Availability"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Availability Percentage")
plt.ylabel("")
plt.show()
```
